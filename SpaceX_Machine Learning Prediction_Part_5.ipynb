This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

```python
#this is my complete code in one block
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Define the parameters
parameters = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [2 * n for n in range(1, 10)],
    'max_features': ['auto', 'sqrt'],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10]
}

# Initialize the Decision Tree Classifier
tree = DecisionTreeClassifier()

# Set up GridSearchCV
tree_cv = GridSearchCV(estimator=tree, param_grid=parameters, cv=10)

# Fit the model on the training data
tree_cv.fit(X_train, Y_train)

# Output the best parameters and the best accuracy score
print("tuned hyperparameters :(best parameters) ", tree_cv.best_params_)
print("accuracy :", tree_cv.best_score_)
```

### Task 12: Model Performance Comparison

We evaluated four classification modelsâ€”**Logistic Regression**, **Support Vector Machine (SVM)**, **Decision Tree**, and **K Nearest Neighbors (KNN)**â€”based on their accuracy on the test data and confusion matrix results.

#### Accuracy Scores:
- **Logistic Regression**: 0.8333
- **SVM**: 0.8333
- **Decision Tree**: **0.8333**
- **K Nearest Neighbors**: 0.8333

#### Best Cross-Validation Accuracy (from GridSearchCV):
- **Logistic Regression**: 0.8464
- **SVM**: 0.8482
- **Decision Tree**: **0.8732**
- **K Nearest Neighbors**: 0.8482

#### Confusion Matrix Summary:
All four models showed the same confusion matrix result:
- True Positive (Landed): 12  
- False Positive: 3  
- False Negative: 0  
- True Negative: 3

#### Conclusion:
While all models performed identically on the test set, the **Decision Tree Classifier** achieved the highest cross-validation accuracy (0.8732). This suggests it generalizes better during model selection and may be the most robust among the four models evaluated.

**âœ… Therefore, the Decision Tree Classifier is considered the best-performing model overall.**


```python
import matplotlib.pyplot as plt

# Model names
models = ['Logistic Regression', 'SVM', 'Decision Tree', 'KNN']

# Accuracies
test_accuracies = [0.8333, 0.8333, 0.8333, 0.8333]
cv_accuracies = [0.8464, 0.8482, 0.8732, 0.8482]

# Bar width and position
bar_width = 0.35
x = range(len(models))

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x, test_accuracies, width=bar_width, label='Test Accuracy')
plt.bar([i + bar_width for i in x], cv_accuracies, width=bar_width, label='CV Accuracy')

# Labels and title
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracies')
plt.xticks([i + bar_width / 2 for i in x], models, rotation=15)
plt.ylim(0.82, 0.88)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```
